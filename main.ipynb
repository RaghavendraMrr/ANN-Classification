```python
# salary prediction
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pickle
```

```python
# read data from CSv
ExcelData=pd.read_csv("Churn_Modelling.csv")
#Remove RowNumber,CustomerID and surname 
data=ExcelData.drop(columns=["RowNumber","CustomerId","Surname"],axis=1)
data
```

```python
# Label for Gender and onehotencoding for Geography
```

```python
label_encoder=LabelEncoder()
data['Gender']=label_encoder.fit_transform(data['Gender'])
data
```

```python
# one hot encoding
onehot_Encoder=OneHotEncoder()
encoded_columns=onehot_Encoder.fit_transform(data[['Geography']])
encoded_df=pd.DataFrame(encoded_columns.toarray(),columns=onehot_Encoder.get_feature_names_out(["Geography"]))
data = pd.concat([data, encoded_df], axis=1)
data = data.drop(columns=["Geography"])
data
```

```python
# split the data into dependent and independent features
x=data.drop('EstimatedSalary',axis=1)
y=data['EstimatedSalary']
```

```python
# Train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# scale the features using StandardScalar
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# Save the encoders and scaler for later use
with open('label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)

with open('onehot_encoder_geo.pkl', 'wb') as file:
    pickle.dump(onehot_Encoder, file)

with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
#model creation.. 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),  # hidden layer
    Dense(1) #output layer
])

model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mae'])
model.summary()
model.save('model.h5')
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime

#set up TensorBoard
log_dir="regressionlogs/fit/"
tensorboard_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# Set up Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
#Training the model
model.fit(x_train,y_train,
          validation_data=(x_test,y_test),
          epochs=100,
          callbacks=[early_stopping_callback,tensorboard_callback]
          )
```
